It includes information about daily activity, steps, and heart rate that can be used to explore users’ habits.


```python
daily_df = pd.concat((pd.read_csv('/kaggle/input/fitbit/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/dailyActivity_merged.csv'),
                      pd.read_csv('/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv')),\
                      axis=0).reset_index(drop=True)
daily_df.head(2)
```

```python
minute_df_dict= {}
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if re.findall('^minute.+Narrow.*',filename):
            if filename in minute_df_dict:
                minute_df_dict[filename] = pd.concat((minute_df_dict[filename],pd.read_csv(os.path.join(dirname, filename))),axis=0)
            else:
                minute_df_dict[filename] = pd.read_csv(os.path.join(dirname, filename))
                
_, minute_df = minute_df_dict.popitem()
for df in minute_df_dict.values():
    minute_df = minute_df.merge(df,on=['Id','ActivityMinute'],how='inner')
minute_df.head(2)
```

A MET is a way to measure your body’s expenditure of energy. The higher the MET value of a particular activity, the more energy your muscles will need to expend to do that activity.

> One MET is approximately 3.5 milliliters of oxygen consumed per kilogram of body weight per minute.

```python
def remove_outliers(df,col):
    Q1= df[col].quantile(.25)
    Q3 = df[col].quantile(.75)
    IQR = Q3 - Q1
    low_lim= Q1 - (1.5 * IQR)
    up_lim = Q3 + (1.5 * IQR)
    return df[ (df[col]>=low_lim) & (df[col]<=up_lim) ]
```

```python
mean_minute_df = minute_df.groupby('minute')[['Intensity','Calories','METs','Steps']].agg('mean')
```

```python
plt.plot(mean_minute_df.index,mean_minute_df.Calories,marker='o',label='Calories')
plt.xlabel('Minute of Hour')
plt.ylabel('Mean Calories per minute of day')
plt.xticks(range(0,61,10))
plt.show()
```

```python
mean_hour_df = minute_df.groupby('hour')[['Intensity','Calories','METs','Steps']].agg('mean')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(mean_hour_df.index,mean_hour_df.Calories,marker='o',label='Calories')
plt.xlabel('Hour of day')
plt.ylabel('Mean Calories per hour of day')

plt.subplot(1,2,2)
plt.plot(mean_hour_df.index,mean_hour_df.METs,marker='o',label='METs')
plt.xlabel('Hour of day')
plt.ylabel('Mean METs per hour of day')
plt.show()
```

```python
mean_day_df = minute_df.groupby('day')[['Intensity','Calories','METs','Steps']].agg('mean')
```

```python
mean_day_df = mean_day_df.reindex( pd.Index(['Mon','Tue', 'Wed', 'Thu','Fri', 'Sat', 'Sun']))
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(mean_day_df.index,mean_day_df.METs,marker='o',label='METs')
plt.ylabel('Mean METs per day of week')

plt.subplot(1,2,2)
plt.plot(mean_day_df.index,mean_day_df.Calories,marker='o',label='Clories')
plt.ylabel('Mean Calories per day of week')
plt.show()
```
